## 🎯 Business Problem

The Human Resources department wants to understand why employees leave the organization.

The primary objectives of this analysis are:

- Understand employee demographics.
- Identify factors influencing attrition.
- Analyze salary, overtime, job satisfaction, and work-life balance.

## 🏷️ Feature Classification

After removing the **EmployeeNumber** column, the remaining variables were grouped based on their data type and business meaning.

### Why was EmployeeNumber removed?

`EmployeeNumber` is a unique identifier assigned to each employee. Since every employee has a different ID, it does not contain any meaningful information that can explain employee attrition or contribute to statistical analysis. Therefore, it was removed before performing EDA.

### Why were the features segregated?

The remaining features were categorized into nominal, ordinal, discrete, and continuous variables based on their data type and business meaning. This classification helps organize the exploratory data analysis and ensures appropriate statistical summaries and visualizations are selected for each variable type.

- **Nominal Variables:** Categorical variables without any natural order.
  - *Examples:* Gender, Department, JobRole, MaritalStatus, BusinessTravel.

- **Ordinal Variables:** Categorical variables with a meaningful ranking or order.
  - *Examples:* Education, JobSatisfaction, EnvironmentSatisfaction, WorkLifeBalance, StockOptionLevel.

- **Discrete Variables:** Numerical variables that take countable integer values.
  - *Examples:* DistanceFromHome, TotalWorkingYears, YearsAtCompany, NumCompaniesWorked.

- **Continuous Variables:** Numerical variables measured over a continuous range.
  - *Examples:* Age, MonthlyIncome, DailyRate, HourlyRate, MonthlyRate.

This classification makes the exploratory data analysis more organized and helps in selecting appropriate statistical summaries and visualizations for each type of variable.

```python
for col in nominal:

    print('='*70)
    print(f'📌 Attrition vs {col}')
    print('='*70)

    # Crosstab (Counts)
    print('\nCounts:\n')
    print(pd.crosstab(df[col], df['Attrition']))

    # Crosstab (Percentages)
    print('\nPercentages:\n')
    print(
        round(
            pd.crosstab(
                df[col],
                df['Attrition'],
                normalize='index'
            ) * 100,
            2
        )
    )

    # Visualization
    plt.figure(figsize=(8,5))

    ax = sns.countplot(
        data=df,
        x=col,
        hue='Attrition'
    )

    for container in ax.containers:
        ax.bar_label(container)

    plt.title(f'Attrition by {col}')
    plt.xlabel(col)
    plt.ylabel('Number of Employees')

    if df[col].nunique() > 3:
        plt.xticks(rotation=20)

    plt.show()
```

# 💼 Business Recommendations

Based on the analysis conducted on the IBM HR Attrition dataset, the following recommendations are proposed to improve employee retention and reduce attrition:

### 1. Reduce Employee Overtime
Employees who work overtime exhibit a considerably higher attrition rate. HR should review workload distribution, encourage work-life balance, and minimize excessive overtime whenever possible.

### 2. Improve Job Satisfaction
Lower job satisfaction is associated with higher employee attrition. Regular employee feedback, recognition programs, career development opportunities, and supportive leadership can help improve overall job satisfaction.

### 3. Strengthen Work-Life Balance
Employees reporting poor work-life balance are more likely to leave the organization. Flexible work arrangements, wellness initiatives, and balanced workloads should be promoted to improve employee well-being.

### 4. Focus on High-Attrition Departments and Job Roles
Departments and job roles with relatively higher attrition should receive targeted attention. HR should conduct exit interviews, identify department-specific challenges, and develop customized retention strategies.

### 5. Improve Early Employee Retention
Employees with fewer years at the company and lower overall work experience show higher attrition. Structured onboarding, mentoring programs, and clear career progression plans can improve retention during the early years of employment.

### 6. Review Compensation Strategies
Employees with lower monthly incomes tend to leave more frequently. Periodic salary benchmarking and performance-based rewards can help maintain competitive compensation and improve employee retention.
